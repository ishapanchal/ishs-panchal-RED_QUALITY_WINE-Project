```python
#The Red wine quality Dataset:

#Involves predicting the onset of quality within Red_Wine given details. It is a binary (2-class) classification problem. 
#The number of observations for each class is not balanced. There are 1599 observations with 11 input variables and 1 output variable.

#The variable names are as follows:

#About the Data set The data set contains 1,599 observations on Red Wines. There are 12 different variables to describe a wine. 
#Each variable is

#Fixed Acidity: most acids involved with wine or fixed or nonvolatile (do not evaporate readily)

#Volatile Acidity: the amount of acetic acid in wine, which at too high of levels can lead to an unpleasant, vinegar taste

#Citric Acid: found in small quantities, citric acid can add 'freshness' and flavor to wines.

#Residual Sugar: the amount of sugar remaining after fermentation stops, it's rare to find wines with less than 1 gram/liter 
#and wines with greater than 45 grams/liter are considered sweet.

#Chlorides: the amount of salt in the wine.

#Free Sulfur Dioxide: the free form of SO2 exists in equilibrium between molecular SO2 (as a dissolved gas) and bisulfite ion; 
#it prevents microbial growth and the oxidation of wine.

#Total Sulfur Dioxide: amount of free and bound forms of S02; in low concentrations, SO2 is mostly undetectable in wine, 
#but at free SO2 concentrations over 50 ppm, SO2 becomes evident in the nose and taste of wine.

#Density: the density of water is close to that of water depending on the percent alcohol and sugar content.

#pH: describes how acidic or basic a wine is on a scale from 0 (very acidic) to 14 (very basic); most wines are between 3-4 
#on the pH scale.

#Sulphates: a wine additive which can contribute to sulfur dioxide gas (S02) levels, which acts as an antimicrobial and 
#antioxidant.

#Alcohol: the percent alcohol content of the wine

#Quality: score between 0 and 10

#Variable classification:

#Quantitative:

#1. Fixed Acidity: most acids involved with wine.

#2. Volatile Acidity: the amount of acetic acid in wine.

#3. Citric Acid: found in small quantities.

#4. Residual Sugar: the amount of sugar remaining after fermentation stops.

#5. Chlorides: the amount of salt in the wine.

#6. Free Sulfur Dioxide: it prevents microbial growth and the oxidation of wine.

#7. Total Sulfur Dioxide: amount of free and bound forms of S02. SO2 becomes evident in the nose and taste of wine.

#8. Density: the density of water is close to that of water depending on the percent alcohol and sugar content.

#9. pH: describes how acidic or basic a wine is on a scale from 0 (very acidic) to 14 (very basic).

#10. Sulphates: a wine additive which can contribute to sulfur dioxide gas (S02) levels.

#11. Alcohol: the percent alcohol content of the wine

#Categorical:

#12. Quality: score between 0 and 10
```

```python
#Using Logisitic Regression to find the accuracy of red wine

#First spilit the data into train and test

from sklearn.model_selection import train_test_split as tts
train, test=tts(red,test_size=0.2,random_state=42)

```

```python
train_x=train.loc[:, train.columns !="quality"]
test_x=test.loc[:, train.columns !="quality"]

train_y=train["quality"]
test_y=test['quality']
```

```python
# making logistic alogithim

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train_x, train_y)
```

```python
#To predict the result

test_pred = lr.predict(test_x)
```

```python
result=pd.DataFrame({'Actual Score':test_y,'predicted_score':test_pred})
print(result)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(test_y,test_pred))
```

```python
print(confusion_matrix(test_y,test_pred))
```

```python
accuracy_score(test_y,test_pred)*100
```

```python
#Using KNN Alogirithm

from sklearn.neighbors import KNeighborsClassifier
kn=KNeighborsClassifier(n_neighbors=3)
kn.fit(train_x,train_y)
```

```python
test_pred1 = kn.predict(test_x)
```

```python
result=pd.DataFrame({'Actual Score':test_y,'predicted_score':test_pred1})
print(result)
```

```python
print(classification_report(test_y,test_pred1))
```

```python
print(confusion_matrix(test_y,test_pred1))
```

```python
accuracy_score(test_y,test_pred1)*100
```
